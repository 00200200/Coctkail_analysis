```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
df_cocktail = pd.read_json('../data/cocktail_dataset.json')
df_cocktail
```

```python
def extract_ingredients(ingredients):
    list_of_ingredients = []
    for ingredient in ingredients:
        list_of_ingredients.append(ingredient['name'])
    
    return list_of_ingredients

df_cocktail['ingredients'] = df_cocktail['ingredients'].apply(extract_ingredients)    
```

```python
category_counts = df_cocktail[['category']].value_counts()
category_counts.plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Number of cocktails')
```

```python
df_cocktail['number_of_ingredients'] = df_cocktail['ingredients'].apply(len)
plt.hist(df_cocktail['number_of_ingredients'])
```

```python
glass_counts = df_cocktail['glass'].value_counts()
glass_counts.plot(kind='bar')
```

```python
ingredients_columns = df_cocktail['ingredients'].str.join("|").str.get_dummies()
tags_columns = df_cocktail['tags'].str.join("|").str.get_dummies()
# glass_columns = df_cocktail['glass'].str.join("|").str.get_dummies()  Ahh
glass_columns = pd.get_dummies(df_cocktail['glass'])
df_cocktail_encoded = pd.concat([df_cocktail,ingredients_columns,glass_columns],axis=1)
df_cocktail_encoded
```

```python
df_cocktail_encoded.drop(['id','name','category','glass','instructions','imageUrl','createdAt','updatedAt','tags','ingredients','alcoholic','number_of_ingredients'],axis=1,inplace=True)
# df_cocktail_encoded_without_alcoholic = df_cocktail_encoded.copy()
# df_cocktail_encoded_without_alcoholic.drop(['alcoholic'],axis=1,inplace=True)
```

```python
df_cocktail_encoded
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans()

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.metrics import davies_bouldin_score
wcss = []
davies_scores = []
silhouette_scores = []
for i in range(2,15):
    kmeans = KMeans(n_clusters=i,random_state=42)
    kmeans.fit(df_cocktail_encoded)
    wcss.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(df_cocktail_encoded,kmeans.labels_))
    labels = kmeans.labels_
    score = davies_bouldin_score(df_cocktail_encoded, kmeans.labels_)
    davies_scores.append(score)

```

```python

plt.figure(figsize=(14,6))


plt.subplot(1, 3, 1)
plt.plot(range(2, 15), wcss, marker='o')
plt.title('Inertia Values for Different Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.grid(True)

plt.subplot(1, 3, 2)
plt.plot(range(2, 15), silhouette_scores, marker='o', color='orange')
plt.title('Silhouette Scores for Different Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.grid(True)

plt.subplot(1, 3,3)
plt.plot(range(2, 15), davies_scores, marker='o', color='green')
plt.title('davies for Different Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('davies Score')
plt.grid(True)
```

```python
optimal_k = 5
kmeans = KMeans(n_clusters=optimal_k,random_state=42)
kmeans.fit(df_cocktail_encoded)
kmeans_labels = kmeans.labels_
```

```python
from sklearn.decomposition import PCA
pca = PCA(2)
df_cocktail_encoded = pca.fit_transform(df_cocktail_encoded)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_cocktail_encoded[:, 0], df_cocktail_encoded[:, 1], c=kmeans_labels, cmap='viridis', s=50)
plt.scatter(pca.transform(kmeans.cluster_centers_)[:, 0], pca.transform(kmeans.cluster_centers_)[:, 1], c='red', marker='x', s=200)
plt.title(f"KMeans Clustering result for k = {optimal_k}")

plt.show()
```

```python
df_cocktail['kmean_labels'] = kmeans_labels
```
